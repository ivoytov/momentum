```python
import pandas as pd
import numpy as np
import pyfolio as pf

from scipy.stats import pearsonr
```

```python
df = pd.read_csv('data.csv', parse_dates=['trading_day'], index_col=[1,0]).unstack('symbol').dropna(axis=1)
```

Separate the data into test and training set

```python
test = df.iloc[len(df.index)//2:]
train = df.iloc[:len(df.index)//2]
```

We measure momentum as the security gaining in price over a `lookback` period in days. If momentum is detected, we buy the security and hold it for a look forward period `holddays` in days. We compute correlations and pvalues of various combinations of periods. We observe that the highest correlation is for a 30 day lookback and 10 day hold, with correlation of `0.18`. Another alternative is `(10, 5)` with correlation of `.1682`. P values for both are zero but given the small data set this may not be significant. We pick the shorter hold period due to its advantage as it offers us more trading opportunities.

```python
# correlation tests
cl = train['close']
print("lookback    holddays   correlation   pvalue")
for lookback in [1, 5, 10, 25, 30]:
    for holddays in [1, 5, 10, 25, 30]:
        ret_lag = cl.pct_change(periods=lookback)
        ret_fut = cl.pct_change(periods=-holddays)
        bad_dates = ret_lag.isna().any(axis=1) | ret_fut.isna().any(axis=1)
        ret_lag.loc[bad_dates] = 0
        ret_fut.loc[bad_dates] = 0
        if lookback >= holddays:
            indepSet = range(1,len(ret_lag)-1,lookback)
        else:
            indepSet = range(1,len(ret_lag)-1,holddays)
            
        ret_lag = ret_lag.iloc[indepSet]
        ret_fut = ret_fut.iloc[indepSet]
        corr, pval = pearsonr(ret_lag.unstack(), ret_fut.unstack())
        

        print(f"  {lookback: >8}   {holddays:>8}  {corr:>8.4f}   {pval:>8.4f}")
```

```python
lookback = 10
holddays = 5
cl = train['close']
longs = cl > cl.shift(lookback)
shorts = cl < cl.shift(lookback)
pos = pd.DataFrame(0, index=cl.index, columns=cl.columns)
for h in range(0, holddays):
    long_lag = longs.shift(h)
    long_lag = long_lag.fillna(False)
    
    short_lag = shorts.shift(h)
    short_lag = short_lag.fillna(False)
    
    # trivial to add shorting if needed - just add a short_lag
    pos[long_lag] += 1
    pos[short_lag] -= 1

ret = pos.shift(1) * cl.pct_change() / holddays
port_ret = ret.sum(axis=1) / len(ret.columns)
print(
    "Sharpe ratio",
    "{:4.2f}".format(pf.timeseries.sharpe_ratio(index_ret)),
)
(port_ret+1).cumprod().plot()
```

```python
index = cl.pct_change()
index_ret = index.sum(axis=1) / len(index.columns)
print(
    "Sharpe ratio",
    "{:4.2f}".format(pf.timeseries.sharpe_ratio(port_ret)),
)
(index_ret+1).cumprod().plot()
```
